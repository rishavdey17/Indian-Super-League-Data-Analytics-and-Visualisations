```python
shot_data = pd.read_csv('ISL Shots.csv')
```

```python
shot_data
```

```python
shot_data_home = shot_data[shot_data['Team'] == 'Home']
shot_data_away = shot_data[shot_data['Team'] == 'Away']
```

```python
shot_data_home
```

```python
shot_data_away
```

```python
home_goal = shot_data_home[shot_data_home["Result"] == 'Goal'].copy()
home_saved = shot_data_home[shot_data_home["Result"] == 'Saved'].copy()
home_miss = shot_data_home[shot_data_home["Result"] == 'Missed'].copy()
home_blocked = shot_data_home[shot_data_home["Result"] == 'Blocked'].copy()
home_woodwork = shot_data_home[shot_data_home["Result"] == 'Woodwork'].copy()

away_goal = shot_data_away[shot_data_away["Result"] == 'Goal'].copy()
away_saved = shot_data_away[shot_data_away["Result"] == 'Saved'].copy()
away_miss = shot_data_away[shot_data_away["Result"] == 'Missed'].copy()
away_blocked = shot_data_away[shot_data_away["Result"] == 'Blocked'].copy()
away_woodwork = shot_data_away[shot_data_away["Result"] == 'Woodwork'].copy()
```

```python
home_goal['x'] = home_goal['x'] * 1.2
home_goal['y'] = home_goal['y'] * 0.8

home_saved['x'] = home_saved['x'] * 1.2
home_saved['y'] = home_saved['y'] * 0.8

home_miss['x'] = home_miss['x'] * 1.2
home_miss['y'] = home_miss['y'] * 0.8

home_blocked['x'] = home_blocked['x'] * 1.2
home_blocked['y'] = home_blocked['y'] * 0.8

home_woodwork['x'] = home_woodwork['x'] * 1.2
home_woodwork['y'] = home_woodwork['y'] * 0.8
```

```python
away_goal['x'] = away_goal['x'] * 1.2
away_goal['y'] = away_goal['y'] * 0.8

away_saved['x'] = away_saved['x'] * 1.2
away_saved['y'] = away_saved['y'] * 0.8

away_miss['x'] = away_miss['x'] * 1.2
away_miss['y'] = away_miss['y'] * 0.8

away_blocked['x'] = away_blocked['x'] * 1.2
away_blocked['y'] = away_blocked['y'] * 0.8

away_woodwork['x'] = away_woodwork['x'] * 1.2
away_woodwork['y'] = away_woodwork['y'] * 0.8
```

```python
pitch = Pitch(pitch_type='statsbomb',
                      pitch_color='#0e171c', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(13.5,8), constrained_layout=True, tight_layout=False)
fig.set_facecolor('w')

pitch.scatter(120-home_goal['x'], home_goal['y'],
              s = home_goal['xG'] * 500 + 30,
              c ='#00ff00', edgecolor = 'white', marker = 'football',
              ax=ax)

pitch.scatter(120-home_saved['x'], home_saved['y'],
              s = home_saved['xG'] * 500 + 30,
              c ='#ffdc2e', edgecolor = 'white',
              ax=ax)

pitch.scatter(120-home_blocked['x'], home_blocked['y'],
              s = home_blocked['xG'] * 500 + 30,
              c ='#dc6601', edgecolor = 'white',
              ax=ax)

pitch.scatter(120-home_woodwork['x'], home_woodwork['y'],
              s = home_woodwork['xG'] * 500 + 30,
              c ='w', edgecolor = 'white',
              ax=ax)

pitch.scatter(120-home_miss['x'], home_miss['y'],
              s = home_miss['xG'] * 500 + 30,
              c ='r', edgecolor = 'white',
              ax=ax)

pitch.scatter(away_goal['x'], away_goal['y'],
              s = away_goal['xG'] * 500 + 30,
              c ='#00ff00', edgecolor = 'white', marker = 'football',
              ax=ax)

pitch.scatter(away_saved['x'], away_saved['y'],
              s = away_saved['xG'] * 500 + 30,
              c ='#ffdc2e', edgecolor = 'white',
              ax=ax)

pitch.scatter(away_blocked['x'], away_blocked['y'],
              s = away_blocked['xG'] * 500 + 30,
              c ='#dc6601', edgecolor = 'white',
              ax=ax)

pitch.scatter(away_woodwork['x'], away_woodwork['y'],
              s = away_woodwork['xG'] * 500 + 30,
              c ='w', edgecolor = 'white',
              ax=ax)

pitch.scatter(away_miss['x'], away_miss['y'],
              s = away_miss['xG'] * 500 + 30,
              c ='r', edgecolor = 'white',
              ax=ax)

plt.title("Shot Map, FC Goa 1-0 East Bengal FC.", fontsize = 18, c = '#000000')

highlight_text.fig_text(x = 0.328, y = 0.94,
                        s = 'FC Goa vs East Bengal FC. Indian Super League MatchDay 17, Margao.',
                        fontsize = 10,
                        color = '#000000'
                       )

plt.scatter(x = 28, y = -2.0, s = 70, c = '#00ff00')
plt.text(x = 29.5, y = -1.2, s = f'Goal.', c = '#00ff00', fontsize = 12)

plt.scatter(x = 38, y = -2, s = 70, c = 'w')
plt.text(x = 39.5, y = -1.2, s = f'Woodwork.', c = 'w', fontsize = 12)

plt.scatter(x = 55, y = -2, s = 70, c = '#ffea00')
plt.text(x = 56.5, y = -1.2, s = f'Saved.', c = '#ffea00', fontsize = 12)

plt.scatter(x = 67, y = -2, s = 70, c = '#dc6601')
plt.text(x = 68.5, y = -1.2, s = f'Blocked.', c = '#dc6601', fontsize = 12)

plt.scatter(x = 81, y = -2, s = 70, c = 'r')
plt.text(x = 82.5, y = -1.2, s = f'Missed.', c = 'r', fontsize = 12)


plt.text(x=22, y=87.5, s ='FC GOA', fontsize = 13, fontweight = 'bold', c = '#000000', ha = 'left')

plt.text(x=1, y=90, s ='SHOTS - 9, ', fontsize = 10, c = '#000000', ha = 'left')

plt.text(x=12.5, y=90, s ='EXPECTED GOALS (xG) - 1.32, ', fontsize = 10, c = '#000000', ha = 'left')

plt.text(x=43, y=90, s ='GOALS - 1.', fontsize = 10, c = '#000000', ha = 'left')

plt.text(x=78.5, y=87.5, s ='EAST BENGAL FC', fontsize = 13, fontweight = 'bold', c = '#000000', ha = 'left')

plt.text(x=63, y=90, s ='SHOTS - 13, ', fontsize = 10, c = '#000000', ha = 'left')

plt.text(x=75.5, y=90, s ='EXPECTED GOALS (xG) - 0.84, ', fontsize = 10, c = '#000000', ha = 'left')

plt.text(x=106, y=90, s ='GOALS - 0.', fontsize = 10, c = '#000000', ha = 'left')
```
